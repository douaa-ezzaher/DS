### **Contexte et objectif**

Cette base vise à prédire la performance des élèves au lycée, à partir de nombreuses variables personnelles, sociales et scolaires.
- Les données proviennent de deux écoles portugaises : Gabriel Pereira (GP) et Mousinho da Silveira (MS).
- Population étudiée : élèves portugais inscrits en Mathématiques et Langue portugaise (de 15 à 22 ans).
- Données collectées via bulletins scolaires et questionnaires.

### **Description des fichiers**

- Deux ensembles de données :
  - student-mat.csv (cours de mathématiques)
  - student-por.csv (cours de langue portugaise)
- Nombre total d’élèves : 649
- Nombre total de variables (caractéristiques) : 30
- Aucun point manquant signalé dans les données

### **Variables principales**

- Démographie : âge, sexe, statut parental, taille de la famille, éducation des parents, profession des parents, type d’adresse (urbaine/rurale)
- Scolarité et quotidien : temps de trajet, temps d’étude, présence de soutien scolaire ou familial, cours particuliers, activités extrascolaires, accès à Internet, qualité des relations familiales, temps libre, sorties, consommation d'alcool, état de santé, nombre d’absences
- Notes scolaires :
  - G1 : note du 1er trimestre
  - G2 : note du 2ème trimestre
  - G3 : note finale (3ème trimestre, cible principale de prédiction)

### **Types de tâches associées**

- Classification binaire, classification à cinq niveaux, et régression
- L’attribut cible G3 (note finale) est très corrélé avec G1 et G2, ce qui reflète le processus d’évaluation au lycée.

### **Publications de référence**

- Plusieurs articles scientifiques ont utilisé ce jeu de données pour l’analyse prédictive, l’intelligence artificielle et la psychométrie.
